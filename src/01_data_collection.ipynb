```python
import pandas as pd
import os

os.makedirs('data', exist_ok=True)

# ── Option A: HuggingFace dataset ──
try:
    from datasets import load_dataset
    dataset  = load_dataset('dair-ai/emotion')
    train_df = pd.DataFrame(dataset['train'])
    test_df  = pd.DataFrame(dataset['test'])
    df = pd.concat([train_df, test_df], ignore_index=True)

    emotion_map = {0:'sadness',1:'joy',2:'love',3:'anger',4:'fear',5:'surprise'}
    df['emotion'] = df['label'].map(emotion_map)
    df = df[['text','emotion']]
    print('✅ Loaded from HuggingFace!')

# ── Option B: Synthetic fallback ──
except Exception as e:
    print(f'⚠️  HuggingFace unavailable ({e}). Using synthetic data.')
    import random; random.seed(42)
    samples = {
        'joy':      ['I am so happy today!','This is the best day ever!','I feel amazing!',
                     'I love spending time with family.','Today I aced my exam!',
                     'Everything is going great.','I just got a promotion!'],
        'sadness':  ['I feel so lonely.','I cannot stop crying.','Nothing seems to matter.',
                     'I miss my old friends.','Life feels really hard.','I feel hopeless.',
                     'I lost someone very close.'],
        'anger':    ['I am furious!','This is so unfair!','I cannot believe they did that.',
                     'I am really frustrated.','Stop lying to me!','I hate this situation.',
                     'This makes my blood boil.'],
        'fear':     ['I am really scared.','Something feels wrong.','I am terrified.',
                     'I feel anxious and cannot sleep.','What if something bad happens?',
                     'I am afraid to face reality.','My heart is pounding.'],
        'surprise': ['I did not expect this!','This is such a shock!','I cannot believe it!',
                     'What a surprise!','I never saw that coming.','Completely unexpected!',
                     'I am astonished.'],
        'love':     ['I love you so much!','You mean the world to me.',
                     'I feel deeply connected to you.','My heart is full of love.',
                     'Being with you makes everything better.','I care about you deeply.',
                     'You are the best thing in my life.']
    }
    rows = [{'text': t, 'emotion': e} for e, texts in samples.items() for t in texts]
    df = pd.DataFrame(rows * 40).sample(frac=1, random_state=42).reset_index(drop=True)

# ── Save ──
df.to_csv('data/final_dataset.csv', index=False)
print(f'\n✅ Saved → data/final_dataset.csv')
print(f'📊 Shape : {df.shape}')
print(f'🏷️  Classes: {df["emotion"].unique().tolist()}')
df.head(8)
```
