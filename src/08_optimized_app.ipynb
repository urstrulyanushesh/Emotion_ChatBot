```python
import sys, random
sys.path.append('utils')
from helpers import load_model_artifacts, chatbot_respond, EMOTION_EMOJIS
import gradio as gr

# ── Load model ──
model, tfidf, le = load_model_artifacts('models/')
print('✅ Models loaded — building Gradio app...')
```

```python
# ── Gradio chat function ──
def respond(message, history):
    reply, emotion, conf = chatbot_respond(message, model, tfidf, le)
    emoji    = EMOTION_EMOJIS.get(emotion, '🤖')
    conf_str = f'{conf:.1%}' if conf else ''
    full_reply = f'{reply}\n\n_{emoji} Detected: **{emotion.upper()}** {conf_str}_'
    history.append((message, full_reply))
    return history, history

# ── Build Gradio UI ──
with gr.Blocks(theme=gr.themes.Soft(), title='Emotion-Aware AI Chatbot') as demo:

    gr.Markdown("""
    # 🤖 Emotion-Aware AI Chatbot
    ### ML Project — AI Class (Tuesday)
    *Share how you feel — I understand your emotions and respond with empathy.*
    ---
    """)

    chatbot = gr.Chatbot(height=420, label='Chat')
    state   = gr.State([])

    with gr.Row():
        msg     = gr.Textbox(placeholder='Type how you feel...', show_label=False, scale=4)
        send_btn = gr.Button('Send 💬', scale=1)

    gr.Examples(
        examples=[
            'I am so happy today, this is the best day!',
            'I feel really sad and alone.',
            'I am absolutely furious right now!',
            'I am terrified about my results.',
            'I love spending time with my family.',
            'Oh wow, I totally did not expect that!'
        ],
        inputs=msg
    )

    send_btn.click(respond, [msg, state], [chatbot, state])
    msg.submit(respond,     [msg, state], [chatbot, state])

    gr.Markdown('---\n*Built with TF-IDF + ML Classification | Emotion-Aware Response Engine*')

print('🚀 Launching Gradio app...')
demo.launch(share=True)   # share=True → generates public link
```
