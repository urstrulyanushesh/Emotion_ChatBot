```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

# ── Load cleaned data ──
df = pd.read_csv('data/cleaned_data.csv')
print(f'✅ Loaded data/cleaned_data.csv → shape: {df.shape}')
df.head(5)
```

```python
# ── Plot 1: Emotion Distribution + Word Count ──
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Emotion Dataset — EDA Overview', fontsize=16, fontweight='bold')

counts = df['emotion'].value_counts()
colors = ['#FF6B6B','#FFD93D','#6BCB77','#4D96FF','#C77DFF','#FF9A3C']
axes[0].bar(counts.index, counts.values, color=colors, edgecolor='white', linewidth=1.5)
axes[0].set_title('Emotion Class Distribution', fontweight='bold')
axes[0].set_xlabel('Emotion'); axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=30)
for i, v in enumerate(counts.values):
    axes[0].text(i, v+5, str(v), ha='center', fontweight='bold')

df.boxplot(column='text_length', by='emotion', ax=axes[1], patch_artist=True)
axes[1].set_title('Word Count by Emotion'); axes[1].set_xlabel('Emotion')
plt.suptitle('')
plt.tight_layout(); plt.show()
```

```python
# ── Plot 2: WordClouds per Emotion ──
emotions  = df['emotion'].unique()
cmaps     = ['Reds','Blues','Greens','Purples','Oranges','YlOrRd']
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Word Clouds by Emotion', fontsize=18, fontweight='bold')

for ax, emo, cmap in zip(axes.flatten(), emotions, cmaps):
    texts = ' '.join(df[df['emotion'] == emo]['cleaned_text'].values)
    wc = WordCloud(width=400, height=250, background_color='white',
                   colormap=cmap, max_words=80).generate(texts)
    ax.imshow(wc, interpolation='bilinear')
    ax.axis('off')
    ax.set_title(f'{emo.upper()}', fontweight='bold', fontsize=13)

plt.tight_layout(); plt.show()
```

```python
# ── Plot 3: Correlation heatmap of text length stats ──
stats = df.groupby('emotion')['text_length'].describe()
print('📊 Text Length Stats per Emotion:')
print(stats.round(2))
```
