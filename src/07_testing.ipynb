# 🧪 Step 8 — Testing: Alpha & Beta
**File:** `07_testing.ipynb`

**Input:**  All saved `models/*.pkl` files  ← notebooks 04 & 05

**Alpha Testing** → Automated unit tests (developer)

**Beta  Testing** → Live interactive terminal chatbot (user-facing)

```python
import sys
sys.path.append('utils')
from helpers import load_model_artifacts, predict_emotion, chatbot_respond

model, tfidf, le = load_model_artifacts('models/')
print('✅ Models loaded for testing!')
```

## 🔬 Alpha Testing — Automated Unit Tests

```python
alpha_tests = [
    {'input': 'I feel absolutely wonderful!',        'expected': 'joy'},
    {'input': 'I am devastated and heartbroken.',    'expected': 'sadness'},
    {'input': 'I cannot believe how angry I am!',   'expected': 'anger'},
    {'input': 'I am scared and terrified!',          'expected': 'fear'},
    {'input': 'Oh my god, I did not expect this!',  'expected': 'surprise'},
    {'input': 'I love you with all my heart.',       'expected': 'love'},
]

print('='*55)
print('   🔬 ALPHA TESTING — Automated Unit Tests')
print('='*55)

passed = 0
for i, test in enumerate(alpha_tests, 1):
    emotion, conf = predict_emotion(test['input'], model, tfidf, le)
    ok     = emotion == test['expected']
    status = '✅ PASS' if ok else f'❌ FAIL (got: {emotion})'
    if ok: passed += 1
    print(f'  Test {i}: {status}')
    print(f'    Input   : {test["input"]}')
    print(f'    Expected: {test["expected"]} | Got: {emotion}\n')

score = passed / len(alpha_tests)
print(f'Alpha Result: {passed}/{len(alpha_tests)} passed ({score:.0%})')
print('🏆 Alpha Test PASSED!' if score >= 0.8 else '⚠️  Needs improvement.')
```

## 💬 Beta Testing — Interactive Terminal Chatbot
Type your message and press **Enter**. Type `quit` to exit.

```python
print('='*55)
print('   💬 BETA TESTING — Interactive Chatbot')
print('='*55)
print('Type how you feel. Type "quit" to exit.\n')

while True:
    user_input = input('You: ').strip()
    if user_input.lower() in ['quit', 'exit', 'q']:
        print('Bot: Goodbye! Take care! 👋')
        break
    reply, emotion, conf = chatbot_respond(user_input, model, tfidf, le)
    conf_str = f'{conf:.1%}' if conf else 'N/A'
    print(f'Bot [{emotion.upper()} | {conf_str}]: {reply}\n')
```
