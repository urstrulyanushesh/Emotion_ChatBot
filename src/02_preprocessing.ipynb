```python
import sys
sys.path.append('utils')          # ← connect to shared helpers
from helpers import clean_text

import pandas as pd

# ── Load from previous notebook ──
df = pd.read_csv('data/final_dataset.csv')
print(f'✅ Loaded data/final_dataset.csv → shape: {df.shape}')
df.head(5)
```

```python
# ── Check quality ──
print('=== Missing Values ===')
print(df.isnull().sum())
print(f'\nDuplicate rows: {df.duplicated().sum()}')

# Drop duplicates & nulls
df = df.dropna().drop_duplicates(subset='text').reset_index(drop=True)
print(f'✅ After cleaning → shape: {df.shape}')
```

```python
# ── Apply text cleaning (from helpers.py) ──
print('🔧 Cleaning text...')
df['cleaned_text'] = df['text'].apply(clean_text)
df['text_length']  = df['cleaned_text'].apply(lambda x: len(x.split()))

# Remove empty rows after cleaning
df = df[df['cleaned_text'].str.strip() != ''].reset_index(drop=True)

print(f'✅ Text cleaned! Shape: {df.shape}')
print(f'📏 Avg word count: {df["text_length"].mean():.1f}')
df[['text','cleaned_text','emotion']].head(8)
```

```python
# ── Save cleaned data for next notebook ──
df.to_csv('data/cleaned_data.csv', index=False)
print('✅ Saved → data/cleaned_data.csv')
print(f'📊 Final shape: {df.shape}')
```
