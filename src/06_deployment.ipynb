```python
import sys
sys.path.append('utils')    # ← shared helpers
from helpers import load_model_artifacts, chatbot_respond, predict_emotion

# ── Load all saved artifacts ──
model, tfidf, le = load_model_artifacts('models/')

print('✅ Model artifacts loaded!')
print(f'   Model    : {type(model).__name__}')
print(f'   Features : {tfidf.max_features}')
print(f'   Classes  : {le.classes_.tolist()}')
```

```python
# ── Test chatbot responses ──
test_inputs = [
    'I am so happy today, everything is perfect!',
    'I feel so sad and lonely right now.',
    'I am absolutely furious about this!',
    'I am terrified about my exam tomorrow.',
    'I love you with all my heart.',
    'Oh wow, I totally did not expect this!'
]

print('💬 Chatbot Deployment Test\n' + '='*50)
for text in test_inputs:
    reply, emotion, conf = chatbot_respond(text, model, tfidf, le)
    conf_str = f'{conf:.1%}' if conf else 'N/A'
    print(f'\n👤 User   : {text}')
    print(f'🎭 Emotion : {emotion.upper()} ({conf_str})')
    print(f'🤖 Bot    : {reply}')
```
